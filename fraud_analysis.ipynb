```python
df["amount"] = df["amount"].fillna(df["amount"].mean())
```

```python
df["country"] = df["country"].fillna("Unknown")
```

```python
df = df.dropna(subset=["customer_id"])
```

```python
df.isnull().sum()
```

```python
import numpy as np
```

```python
df["risk_score"] = np.where(df["amount"] > 5000, "high",
                            np.where(df["amount"] > 1000, "medium", "low"))
```

```python
df
```

```python
risk_score_count = df["risk_score"].value_counts()
```

```python
risk_score_count
```

```python
print(df.groupby("country")["amount"].mean())
```

```python
df.to_csv("cleaned_transactions.csv", index=False)
print("File saved successfully")
```
